```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from time import sleep
import pandas as pd
```

Para realizar o scraping, utilizamos o site Brand24, que centraliza publicações de várias redes sociais e sites com base em hashtags e palavras-chave. Essa plataforma nos possibilitou coletar postagens do TikTok, Twitter, YouTube e outros sites relevantes. No entanto, é importante mencionar que o site nos permite utilizar apenas 10 palavras-chave para iniciar a busca.

Dessa forma, escolhemos as seguintes palavras-chave para o estudo relacionado às festas juninas: #festajunina, festa junina, camisa xadrez, pamonha, curau de milho, quentão, dança de quadrilha, #saojoao e sao joao. Essas palavras-chave nos ajudarão a identificar o conteúdo mais relevante e apropriado para nossos anúncios.

```python
driver = webdriver.Chrome()
driver.get("https://app.brand24.com/panel/results/1053724988?p=1&or=0&cdt=days&dr=w&va=1&d1=2023-05-01&d2=2023-05-26")
```

```python
b = driver.find_element(by= By.XPATH , value= '/html/body/div[1]/div[3]/main/div/div[3]/div/div/div[2]/div[1]/div/div[35]/div/div/nav/ul/li[7]/button')
b = int(b.text)
b
```

```python
texto =[]
rede_social = []
sentimento = []
data =[]
for x in range (1, b + 1):
    driver.get(f"https://app.brand24.com/panel/results/1072781956?p={x}&or=0&cdt=days&dr=4&va=1&d1=2023-05-23&d2=2023-06-22")
    sleep(10)
    for i in range(4,34):
        postagem = driver.find_element(by= By.XPATH , value= f'/html/body/div[1]/div[3]/main/div/div[3]/div/div/div[2]/div[1]/div/div[{i}]/div/div/a')
        rede= driver.find_element(by= By.XPATH , value= f'/html/body/div[1]/div[3]/main/div/div[3]/div/div/div[2]/div[1]/div/div[{i}]/div/header/div[2]/div[2]/div/ul/li[1]')
        dia=  driver.find_element(by= By.XPATH , value= f'/html/body/div[1]/div[3]/main/div/div[3]/div/div/div[2]/div[1]/div/div[{i}]/div/header/div[2]/div[2]/ul/li')
        emocao= driver.find_element(by= By.XPATH , value= f'/html/body/div[1]/div[3]/main/div/div[3]/div/div/div[2]/div[1]/div/div[{i}]/div/header/div[3]/div/span')
        rede_txt = rede.text
        dia_txt = dia.text
        emocao_txt = emocao.text
        rede_social.append(rede_txt)
        sentimento.append(emocao_txt)
        data.append(dia_txt)
        link= postagem.get_attribute('href')
        text= postagem.text
        texto.append(text) 
```

```python
dados = {'Texto': texto , 'Data': data, 'Redes Sociais': rede_social, 'sentimento': sentimento}
df = pd.DataFrame(dados)
df
```

```python
df.to_csv('textos_redes_sociais_novo.csv', index=False)
```
